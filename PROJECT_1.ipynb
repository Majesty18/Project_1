### Renaming  Columns

```python
df = df.rename(columns = {'fixed acidity' : 'fixed_acidity'})
df = df.rename(columns = {'citric acid' : 'citric_acid'})
df = df.rename(columns = {'free sulfur dioxide' : 'free_sulfur_dioxide'})
df = df.rename(columns = {'total sulfur dioxide' : 'total_sulfur_dioxide'})
```

```python
df
```

### Filling Null Values With Mean

```python
mean_Cl = df.chlorides.mean()
mean_Fa = df.fixed_acidity.mean()
mean_Ca = df.citric_acid.mean()
mean_FSD= df.free_sulfur_dioxide.mean()
mean_TSD = df.total_sulfur_dioxide.mean()
mean_D = df.density.mean()
mean_P = df.pH.mean()
mean_Q = df.quality.mean()
means = {'mean_Cl':mean_Cl,'mean_Fa':mean_Fa,'mean_Ca':mean_Ca,'mean_FSD':mean_FSD,'mean_TSD':mean_TSD,
         'mean_D':mean_D,'mean_P':mean_P,'mean_Q':mean_Q}
means
```

```python
print(f'chlorides:{mean_Cl}')
print(f'fixed_acidity:{mean_Fa}')
print(f'citric_acid:{mean_Ca}')
print(f'free_sulfur_dioxide:{mean_FSD}')
print(f'total_sulfur_dioxide:{mean_TSD}')
print(f'density:{mean_D}')
print(f'pH:{mean_P}')
print(f'quality:{mean_Q}')
```

```python
mean_Cl
```

```python
df['chlorides']=df['chlorides'].fillna(mean_Cl) 
df['fixed_acidity']=df['fixed_acidity'].fillna(mean_Fa) 
df['citric_acid']=df['citric_acid'].fillna(mean_Ca) 
df['free_sulfur_dioxide']=df['free_sulfur_dioxide'].fillna(mean_FSD) 
df['total_sulfur_dioxide']=df['total_sulfur_dioxide'].fillna(mean_TSD) 
df['density']=df['density'].fillna(mean_D) 
df['pH']=df['pH'].fillna(mean_P) 
df['quality']=df['quality'].fillna(mean_Q)
```

## PIE CHART OF MEAN VALUES

```python
plt.figure(figsize=(8,10))
plt.pie(means.values(),labels=means.keys(),startangle=0,explode=[0,0,0,0.1,0,0.2,0,0.1],shadow=True,colors=['black','purple','yellow','green','red','pink','blue','brown'],autopct="%2.1f%%")
plt.legend(title="Mean values")
plt.show()
```
